# Next Steps

You now have foothold in the world of python. 

However, you will need to do some practicing in order to feel comfortable. An excellent opportunity for practice are the companion notebooks in the `practice-problems` folder in this repository. Each notebook has a few problems that should be relatively straightforward and a couple of challenge problems. Solutions to these problems are also provided as a reference.

Once you work through those problems, you can find more by doing a web search for practice problems. The internet has plenty of programming practice content, although it can take a little time to find the resource that works best for you.

Finally there are a few bonus optional notebooks for your edification. They are not <i>necessary</i> to understand the boot camp material, but they do expand upon the content presented in the previous lectures. For example, if you have been confused at any mention of the word "object", then you may want to check them out.
